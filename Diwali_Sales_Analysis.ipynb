***Exploratory Data Analysis (EDA)***

Analysis of total sales and purchases based on **Gender**

```python
plot = sb.countplot(x = 'Gender', data = df)
```

```python
plot = sb.countplot(x = 'Gender', data = df)

for bars in plot.containers:
  plot.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
Tpurchase = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sb.barplot(x = 'Gender', y = 'Amount', data = Tpurchase)
```

*From the graph above, we can conclude that most purchases and total sales are made by women and that women's purchasing power is greater than that of men*

Analysis of total purchases based on **Age Groups**

```python
plot = sb.countplot(x = 'Age Group', data = df, hue = 'Gender')
```

```python
plot = sb.countplot(x = 'Age Group', data = df, hue = 'Gender')

for bars in plot.containers:
  plot.bar_label(bars)
```

```python
Tpurchase = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sb.barplot(x = 'Age Group', y = 'Amount', data = Tpurchase)
```

*From the chart above, we can conclude that most shoppers are in the 26-35 age group*

Analysis of total purchases and orders based on **State**

```python
Tpurchase = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)

sb.set(rc = {'figure.figsize':(28,5)})
sb.barplot(x = 'State', y = 'Orders', data = Tpurchase)
```

```python
Tpurchase = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sb.set(rc = {'figure.figsize':(28,5)})
sb.barplot(x = 'State', y = 'Amount', data = Tpurchase)
```

*From the above graph we can conclude that most of the orders and total sales are from Uttar Pradesh, Maharashtra and Karnataka respectively*

Analysis of total sales and purchases based on **Marital Status**

```python
plot = sb.countplot(x = 'Marital_Status', data = df)

sb.set(rc = {'figure.figsize':(5,5)})
for bars in plot.containers:
  plot.bar_label(bars)
```

```python
Tpurchase = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sb.set(rc = {'figure.figsize':(5,5)})
sb.barplot(x = 'Marital_Status', y = 'Amount', data = Tpurchase, hue = 'Gender')
```

*From the graph above, we can conclude that most purchases and total sales are made by married women and that their purchasing power is greater*

Analysis of total sales and purchases based on **Occupation**

```python
plot = sb.countplot(x = 'Occupation', data = df)

sb.set(rc = {'figure.figsize':(20,5)})
for bars in plot.containers:
  plot.bar_label(bars)
```

```python
Tpurchase = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sb.set(rc = {'figure.figsize':(20,5)})
sb.barplot(x = 'Occupation', y = 'Amount', data = Tpurchase)
```

*From the above graph we can conclude that most purchases and total sales are made by the buyers working in IT, Healthcare and Aviation Sector*

Analysis of total sales and purchases based on **Product Category**

```python
plot = sb.countplot(x = 'Product_Category', data = df)

sb.set(rc = {'figure.figsize':(28,5)})
for bars in plot.containers:
  plot.bar_label(bars)
```

```python
Tpurchase = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head()

sb.set(rc = {'figure.figsize':(15,5)})
sb.barplot(x = 'Product_Category', y = 'Amount', data = Tpurchase)
```

*From the above graph we can conclude that most purchases are from Clothing & Apparel, Food and Electronics & Gadgets category respectively and most total sales are from the Food, Clothing & Apparel and Electronics & Gadgets category respectively*

Analysis of total purchases based on **Product ID**

```python
Tpurchase = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head()

sb.set(rc = {'figure.figsize':(10,5)})
sb.barplot(x = 'Product_ID', y = 'Orders', data = Tpurchase)
```

*From the above graph we can conclude that the most purchased purchased is with the Product ID P00265242*

**Conclusion:**



***Married women from the age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation sector are more likely to buy products from Food, Clothing & Apparel and Electronics and Gadgets category.***
